```python
os模块是与操作系统交互的一个接口
查看
# print(os.listdir('.'))  # 列出指定目录下的所有文件和子目录，包括隐藏文件，并以列表方式打印
print(os.stat('new_dir_name'))  # 获取文件/目录信息
print(os.stat('01collections.py'))

创建
# os.mkdir('dir') # 生成单级目录
# os.makedirs('dir1/dir2')    # 生成多层递归目录

删除
# os.removedirs('dir1/dir2')  # 若目录为空，则删除，并递归到上一级目录，如若也为空，则删除，依此类推
# os.rmdir('dir1')    # 删除单级空目录，若目录不为空则无法删除，报错；相当于shell中rmdir dirname
# os.remove('dir1/1.py')    # 删除一个文件

修改
# os.rename('dir1', 'new_dir_name')   # 重命名文件/目录

命令相关
# os.system('ls -al') # 运行shell命令，直接显示结果
ret = os.popen('ls -al')  # 运行shell命令，获取执行结果
print(ret.read())

print('当前目录', os.getcwd())  # 获取当前工作目录，即当前python脚本工作的目录路径
os.chdir('new_dir_name')  # 改变当前脚本工作目录；相当于shell下cd
print(os.getcwd())

path相关
print(os.path.abspath('../xxx.py'))  # 返回path规范化的绝对路径
print(os.path.split(os.getcwd()))  # 将path分割成目录和文件名二元组返回
print(os.path.dirname('/Users/Desktop/python_review/review_module/xxx.py'))  # 返回path的目录。其实就是os.path.split(path)的第一个元素
print(os.path.basename('/hello/ok/hello_world.py'))  # 返回path最后的文件名。如果path以／或\结尾，那么就会返回空值。即os.path.split(path)的第二个元素
print(os.path.exists('/Users'))  # 如果path存在，返回True；如果path不存在，返回False
print(os.path.isabs('/Users/Desktop/python_review/review_module/xxx.py'))
print(os.path.isfile('../05os.py'))  # True  解释：如果path是一个存在的文件，返回True。否则返回False
print(os.path.isdir('../new_dir_name'))  # 如果path是一个存在的目录，则返回True。否则返回False
print(os.path.join(os.getcwd(), 'hello', 'okk'))  # 将多个路径组合后返回，第一个绝对路径之前的参数将被忽略
print(os.path.getatime('../new_dir_name'))  # 返回path所指向的文件或者目录的最后访问时间
print(os.path.getmtime('../new_dir_name'))  # 返回path所指向的文件或者目录的最后修改时间
print(os.path.getsize('../new_dir_name'))  # 返回path的硬盘占用空间，单位:B
```
